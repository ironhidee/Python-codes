```python
import pandas as pd
import numpy as np
import xgboost as xgb
from sklearn import model_selection
```

```python
from sklearn.preprocessing import LabelEncoder
number = LabelEncoder()
```

```python
train = pd.io.parsers.read_csv('data_812513.csv')
```

```python
train.describe()
```

```python
train1 = train.drop(['DATETIME','IGVA_VAL0','IGVB_VAL0','VBVA_VAL0','VBVB_VAL0','VBVB_VAL0_FLAG','Engine'] , axis = 1)
```

```python
train1.ALARM.fillna(0, inplace=True)
train1 = pd.get_dummies(data = train1, columns = ['Prior Shutdown','Event','Failure'])
```

```python
y = train1.ALARM.ravel()
```

```python
train1 = train1.drop('ALARM', axis = 1)
```

```python
x = train1.values
```

```python
dtrain = xgb.DMatrix(x, label = y)
```

```python
params = {'max_depth':10, 'eta':1, 'objective':'multi:softprob', 'num_class': 2, 'min_child_weight':1.2, 'subsample': 0.5, 'colsample_bytree' :0.3}
```

```python
gb = xgb.cv(params, dtrain, num_boost_round=10, nfold=5)
```
