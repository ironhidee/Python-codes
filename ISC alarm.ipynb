```python
import pandas as pd
import numpy as np
import xgboost as xgb
from sklearn import model_selection
```

```python
from sklearn.preprocessing import LabelEncoder
number = LabelEncoder()
```

```python
train = pd.io.parsers.read_csv('data_812513.csv')
```

```python
train.describe()
```

```python
train.head()
```

```python
train1 = train.drop('DATETIME', axis = 1)
```

```python
train1.ALARM.fillna(0, inplace=True)
```

```python
y = train1.ALARM.ravel()
```

```python
train1 = train1.drop('ALARM', axis = 1)
```

```python
x = train1.values
```

```python
dtrain = xgb.DMatrix(x, label = y)
```

```python
params = {'max_depth':2, 'eta':1, 'objective':'multi:softmax', 'num_class': 2}
```

```python
gb = xgb.cv(params, dtrain, num_boost_round=10, nfold=3)
```
